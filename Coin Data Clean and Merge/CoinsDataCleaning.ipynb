```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
df = pd.read_csv("crypto_tradinds.csv")
```

```python
df.head()
```

```python
df.dtypes
```

```python
to_drop = ["market_cap","capitalization_change_1_day","crypto_name","crypto_type","max_supply","site_url","github_url","minable","platform_name","industry_name"]
df.drop(to_drop, inplace=True, axis=1)
```

```python
df.head()
```

```python
df["trade_date"] = pd.to_datetime(df["trade_date"])
```

```python
start = datetime(2020, 1, 20)
df = df[df["trade_date"] > start]
```
