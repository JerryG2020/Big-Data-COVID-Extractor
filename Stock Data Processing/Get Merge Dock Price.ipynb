```python
import os
import sys
import csv
import pandas as pd
import numpy as np
import string
import time
```

```python
# get the date and close price for stock A
dataset = pd.read_csv('A.csv', usecols=[0,4], engine='python', header = 0)

# only get the data from Jan 21st, 2020 to April 1st, 2020
dataset = dataset.iloc[-51:]

# change the name of price column, because we want to join different stocks to one table based on date
dataset = dataset.rename(columns = {'Close': 'A'})

# start with variable result
result = dataset
```

```python
# get the dataset path
print(os.getcwd())
path = os.getcwd()
```

```python
# compute time
time_start = time.time()
for file in os.listdir(path):
    # read all csv files in this directory. i.e. read all stock data one by one
    if file[-4:] == ".csv":
        # get stock name
        filename = file.replace(".csv", "")
        if filename != 'A':
            # get the date and close price for current stock
            temp = pd.read_csv(filename + '.csv', usecols=[0,4], engine='python', header = 0)
            # change the name of price column
            temp = temp.rename(columns = {'Close': filename})
            # some stock has no enough data from Jan 21st, 2020 to April 1st, 2020, we filter it.
            if pd.merge(result, temp, on = 'Date').shape[0] < 51:
                continue
            # get the merge
            result = pd.merge(result, temp, on = 'Date')
time_end = time.time()
#it will take about 2 hours, because the larger the dataframe, the processing time will be longer
print("time used:"time_end - time_start)
```

```python
# store the combined dataframe to csv file
result.to_csv("merged stock price.csv", index = 0)
```
